```python
raw_weather = raw_weather['Date,Wind Direction,Wind Speed (m/s)'.split(',')]
raw_weather.columns = 'date,wind_direction,wind_speed'.split(',')
raw_weather.date = pd.to_datetime(raw_weather.date)
raw_weather.head()
```

```python
sorted(list(set(raw_weather.wind_direction)))
```

```python
# note that the wind direction is the direction from which the wind is blowing
# convert this to destination for easier time vectorizing
```

```python
raw_weather.wind_direction = raw_weather.wind_direction.apply(lambda x: x-180.0 if x >= 180.0 else x+180)
```
