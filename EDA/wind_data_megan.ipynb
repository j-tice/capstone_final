```python
import datetime
datetime.datetime.strptime
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('Meteorological Data.xlsx')
data
```

```python
data = data.drop(columns = ["Unnamed: 3"])
data
```

```python
data["Elevation (m)"] = 270.0
data
```

```python
april_data = data[data['Date'].dt.strftime('%Y-%m') == '2016-04']
august_data = data[data['Date'].dt.strftime('%Y-%m') == '2016-08']
```

```python
n = 240
wind_speed = april_data['Wind Speed (m/s)']
wind_dir = april_data['Wind Direction']
time = april_data['Date']
Y = wind_speed

U = np.cos(wind_dir) * wind_speed
V = wind_speed

plt.figure(figsize=(20,10))
plt.ylim([0, 7])
plt.quiver(time, Y, U, V)
```
