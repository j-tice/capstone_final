## Calculating Significance Cut-Offs

```python
# Cutoff for Sensor 1

mean_sensor1 = sensor_1['Reading'].mean().round(2)
std_sensor1 = sensor_1['Reading'].std().round(2)
cutoff_sensor1 = mean_sensor1 + (3 * std_sensor1).round(2)
```

```python
# Cutoff for Sensor 2

mean_sensor2 = sensor_2['Reading'].mean().round(2)
std_sensor2 = sensor_2['Reading'].std().round(2)
cutoff_sensor2 = mean_sensor2 + (3 * std_sensor2).round(2)
```

```python
# Table of Cut-offs

table = [['Sensor', 'Mean Reading', 'Standard Deviation', 'Cut-Off'],  
         ['Sensor 1', mean_sensor1, std_sensor1, cutoff_sensor1], 
         ['Sensor 2', mean_sensor2, std_sensor2, cutoff_sensor2], 
         ['Sensor 3', mean_sensor3, std_sensor3, cutoff_sensor3], 
         ['Sensor 4', mean_sensor4, std_sensor4, cutoff_sensor4], 
         ['Sensor 5', mean_sensor5, std_sensor5, cutoff_sensor5], 
         ['Sensor 6', mean_sensor6, std_sensor6, cutoff_sensor6], 
         ['Sensor 7', mean_sensor7, std_sensor7, cutoff_sensor7], 
         ['Sensor 8', mean_sensor8, std_sensor8, cutoff_sensor8], 
         ['Sensor 9', mean_sensor9, std_sensor9, cutoff_sensor9]]

print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))
```
